```python

played_salary=fifa[["short_name","wage_eur"]]
```

```python
played_Salarys=played_salary.sort_values(by=["wage_eur"],ascending=False).head()
played_Salarys
```

```python
# checking the players who is nationality is  germany
```

```python
germany=fifa[fifa["nationality"]=="Germany"][0:11]
germany
```

```python
# we need to find the tallest person in the  germany
```

```python
f=germany.sort_values(by=["height_cm"],ascending=False)
```

```python
import matplotlib.pyplot as plt

#
top10_germany = germany.sort_values(by='height_cm', ascending=False).head(10)

# Plot
plt.figure(figsize=(10, 6)) 
plt.bar(top10_germany['short_name'], top10_germany['height_cm'])
plt.xlabel('Player Names')
plt.ylabel('Height (cm)')
plt.title('Top 10 Tallest German FIFA Players')
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()  

# Show the plot
plt.show()
```
